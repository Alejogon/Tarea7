```python
import numpy as np
    
%pylab inline
```

```python
n_x=100
n_t=50
c=1
x = linspace(0, 2.0, n_x)
dx = x[1]-x[0]
dt=0.2*dx
b=c*dt/dx

u= 3*sin(pi*x)

plot(x,u)

```

```python
u_maspast=np.zeros(n_x)

u_past=np.zeros(n_x)

for i in range(n_t):
    u_maspast=u_past.copy()
    u_past=u.copy()
    u_past2=(u_past)**2
    for j in range(0,n_x-1):
        u[i]=u_maspast[i]-(b*(u_past2[j+1]-u_past[j-1])/2.0)
        

plot(x,u)
```
