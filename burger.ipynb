```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt

import numpy as np
```

```python
n_x = 200
n_t = 1500

c = 1
dx = 1.0 / 100.0
dt = 1.0 / 5000.0

B = c*dt / dx

x = linspace(0, 2.0, n_x)
t=linspace(0,1.5,n_t)

print B
```

```python
#Condiciones iniciales 
u = 3.0 * sin(pi*x)
```

```python
plot (x,u)
```

```python
#Condiciones de contorno fijas
u[0]= 0
u[-1] = 0
```

# Primer punto

```python




u_past = zeros(n_x)
u_2past = zeros(n_x)

utodos=np.zeros((n_x,n_t))


for j in range (n_x):
    utodos[j,0]=u[j]


for n in range(n_t):  
    plot(x,u)
    u_past = u.copy() 
    u_2past = u_past.copy()
    for i in range(2,n_x-1):
        u[i]=u_2past[i]-(B/2)*( u_past[i+1]**2 -  u_past[i-1]**2 )
        utodos[i,j]=u[i]
        
```

```python
plot (x,u)#Solucion para t =1.5 dado que es el ultimo valor evaluado
```

```python
#Grafica 3D

x = linspace(0,2,n_x)
t = linspace(0,1.5,n_t)


fig = plt.figure(figsize(15,10))
ax = fig.gca(projection='3d')
x ,t = np.meshgrid(x,t)
surf = ax.plot_wireframe(t,x,utodos)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
```
