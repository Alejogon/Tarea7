```python
n_x = 200
n_t = 1000

c = 1
dx = 1.0 / 100.0
dt = 1.0 / 5000.0

B = c*dt / dx

x = linspace(0, 2.0, n_x)

print B
```

```python
#Condiciones iniciales 
u = 3.0 * sin(pi*x)
```

```python
t = zeros(n_t)
for i in range (n_t):
    t[i] = (i+1)*dt
```

```python
plot (x,u)
```

```python
#Condiciones de contorno fijas
u[0]= 0
u[-1] = 0
```

# Primer punto

```python
u_past = zeros(n_x)

for n in range(n_t):  
    plot(x,u)
    u_past = u.copy() 
    
    for i in range(1,n_x-1):
        u[i]=u_past[i]-(B/2)*( u_past[i+1]**2 -  u_past[i-1]**2 )
        
```

```python
plot (x,u)
```

# Segundo punto

```python
u = 3.0 * sin(pi*x)
u_past = zeros(n_x)

for n in range(n_t):  
    plot(x,u)
    u_past = u.copy() 
    
    for i in range(1,n_x-1):
        u[i]= u_past[i]-(B/4.0)*(u_past[i+1]**2 - u_past[i-1]**2) + ((B**2)/8.0)*((u_past[i+1] + u_past[i])*(u_past[i+1]**2  -  u_past[i]**2) - ( u_past[i] + u_past[i-1])*( u_past[i]**2 - u_past[i-1]**2 ))
        
```

```python
plot (x,u)
```
